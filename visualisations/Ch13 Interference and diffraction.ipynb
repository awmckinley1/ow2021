```python
# Space the 'points'
number_points = 10

point_pos = np.linspace(1,10,number_points)
# Wavefront separation
wavefront = 2
zeros = np.zeros(number_points,dtype=float)

arc = patches.Arc((0,3),2*wavefront,2*wavefront,theta1 = -50, theta2 = 50,linewidth=2,edgecolor='k')
# Basic plot

fig, ax = plt.subplots(figsize=(7,5))

ax.add_patch(arc)
ax.set_aspect('equal')
ax.plot(zeros,point_pos,marker='o',linestyle='none')
ax.plot(zeros+1*wavefront,point_pos,marker='o',linestyle='none')
ax.axis('off')
plt.show()
```

```python
number_points = 6

number_wavefronts = 3

wave_sep = 2

fig, ax = plt.subplots(figsize=(7,7))
ax.set_aspect('equal')

colors=['k','r','b']


for i in np.arange(number_wavefronts):
    ax.plot((wave_sep*i,wave_sep*i),(-1,number_points),linestyle='--',color=colors[i])
    for n in np.arange(number_points):
        ax.plot(wave_sep*i,n,marker='o',linestyle='none',color='k')
        arc = patches.Arc((wave_sep*i,n),2*wavefront,2*wavefront,theta1 = -50, theta2 = 50,linewidth=1,edgecolor=colors[i])
        ax.add_patch(arc)




ax.axis('off')

plt.tight_layout()
#plt.ylim((-4,4))
plt.savefig("ch13-huygenswavelets1.svg",transparent=True)
plt.savefig("ch13-huygenswavelets1.pdf",transparent=True)
plt.show()
```

```python
number_points = 9
number_wavefronts = 3
wave_sep = 2
angle_spread_deg = 40

angles=np.linspace(-angle_spread_deg,angle_spread_deg,number_points)




fig, ax = plt.subplots(figsize=(7,7))
ax.set_aspect('equal')

colors=['k','r','b']


arc1 = patches.Arc((0,0),2*(1)*wave_sep,2*(1)*wave_sep,theta1 = -50, theta2 = 50,linewidth=2,edgecolor='k')
arc2 = patches.Arc((0,0),2*(3)*wave_sep,2*(3)*wave_sep,theta1 = -50, theta2 = 50,linewidth=2,edgecolor='k')


#ax.add_patch(arc1)

#ax.plot(zeros,point_pos,marker='o',linestyle='none')
#ax.plot(zeros+1*wavefront,point_pos,marker='o',linestyle='none')
#ax.plot(x_coords_0,y_coords_0,marker='o',linestyle='none')

for n in np.arange(number_wavefronts):
    x_coords_0 = 2*(n+1)*np.cos(np.radians(angles))
    y_coords_0 = 2*(n+1)*np.sin(np.radians(angles))
    ax.plot(x_coords_0,y_coords_0,marker='o',linestyle='none',color='b')
    arc_n = patches.Arc((0,0),(2*n+2)*wave_sep,(2*n+2)*wave_sep,theta1 = -50, theta2 = 50,linewidth=2,edgecolor='k')
    ax.add_patch(arc_n)
    

arc_deg = patches.Arc((4,0),
                      2*(1)*wave_sep,2*(1)*wave_sep,
                      theta1 = -40, theta2 = 40,linewidth=2,edgecolor='r',linestyle='--')

ax.add_patch(arc_deg)

for angle in angles[0::2]:
    arc_stack = patches.Arc((wave_sep*np.cos(np.radians(angle)),-wave_sep*np.sin(np.radians(angle))),
                            2*(1)*wave_sep,2*(1)*wave_sep,
                            theta1 = (-70-angle), theta2 = (70-angle),
                            linewidth=2,edgecolor='r',linestyle='--')
    ax.add_patch(arc_stack)
    
    
ax.set_xlim(-2,10)
#ax.set_ylim(-6,6)
ax.axis('off')

plt.tight_layout()
#plt.ylim((-4,4))
plt.savefig("ch13-huygenswavelets2.svg",transparent=True)
plt.savefig("ch13-huygenswavelets2.pdf",transparent=True)
plt.show()

```

## Plotting intensity of waves on screen from wide slit.



```python
x_vals = np.linspace(-0.01,.01,100)
a = 0.167E-3 # Slit width
L = 1 # Distance to screen
wavelength = 500e-9 # wavelength of light
size_of_font = 15

fig, ax = plt.subplots(figsize=(7,5))

k = a/(L*wavelength)


#intensity = (np.sinc((np.pi*a*x_vals) / (wavelength*L)))**1

#intensity = (np.sinc(k*x_vals))**1 # np.sinc includes pi

intensity2 = (np.sin(np.pi*k*x_vals))/(np.pi*k*x_vals)
             
        
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
#ax.spines["left"].set_position(("data",0))
ax.spines["bottom"].set_position(("data",0))

ax.set_ylabel('Intensity',fontsize=size_of_font)
ax.set_xlabel('$X$',fontsize=size_of_font)

for n in np.arange(4)[1:]:
    ax.plot(((n*wavelength*L)/a),0,marker='o',color='r')
    ax.text(((n*wavelength*L)/a),0.1,'$\dfrac{'+str(n)+'\lambda L}{a}$',ha='center',fontsize=size_of_font)


#ax.plot(x_vals,intensity**2,color='r')
ax.plot(x_vals,intensity2**2,color='b')

plt.savefig('ch13-wideslitintensity1.pdf',transparent=True)
plt.savefig('ch13-wideslitintensity1.svg',transparent=True)

plt.show()
```
