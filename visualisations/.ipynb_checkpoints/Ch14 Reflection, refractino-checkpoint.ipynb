```python
%config InlineBackend.figure_format = 'retina'
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mplines
import matplotlib
import matplotlib.patches as patches


from matplotlib import animation, rc
from IPython.display import HTML, Image

rc('animation', html='html5') # Needed for in-line movie
```

# Animating reflection refraction

```python

size_of_font = 15

limit=3
points = 10
sample = 2

point_of_reflection_1 = np.linspace(-limit,limit,points)
point_of_reflection_2 = np.flip(point_of_reflection_1)
point_of_reflection = np.concatenate([point_of_reflection_1,point_of_reflection_2])

P_x = point_of_reflection[sample]

point_coord_x = 3
point_coord_y = 2
point_A = (-2,2)
point_A_prime = (-2,-2)
point_B = (2,2)


plt.figure(figsize=(10,10))

#plt.axes().set_aspect('equal')


ax = plt.axes()

rect = patches.Rectangle((-2*limit,-limit), 4*limit, limit, linewidth=1, edgecolor='k', facecolor='b',alpha=0.1)
ax.add_patch(rect)

plt.plot(-point_coord_x,point_coord_y,'o',color='blue',markersize=5)
plt.plot(-point_coord_x,-point_coord_y,'o',color='blue',markersize=5)
plt.plot(point_coord_x,point_coord_y,'o',color='orange',markersize=5)


ax.annotate("$A$", xy=(-1.1*point_coord_x,1.1*point_coord_y),fontsize=size_of_font, color='b')
ax.annotate("$A'$", xy=(-1.2*point_coord_x,-1.2*point_coord_y),fontsize=size_of_font, color='b')
ax.annotate("$B$", xy=(1.1*point_coord_x,1.1*point_coord_y),fontsize=size_of_font, color='orange')

ax.annotate("$\\theta_i$", xy=(P_x-0.4,0.5),fontsize=size_of_font, color='red')
ax.annotate("$\\theta_r$", xy=(P_x+0.1,0.5),fontsize=size_of_font, color='red')




plt.plot((-point_coord_x,P_x),(point_coord_y,0),color='blue',lw=3)
plt.plot((P_x,point_coord_x),(0,point_coord_y),color='blue',lw=3)
plt.plot((-point_coord_x,P_x),(-point_coord_y,0),color='blue',lw=3,linestyle='dotted')
plt.plot((-point_coord_x,point_coord_x),(-point_coord_y,point_coord_y),color='orange',lw=3,linestyle='dotted')



ax.axvline(x=P_x,linewidth=1,color='k',linestyle='dotted')


#plt.plot((x1,x2),(y1,y2)  )

ax.set_aspect('equal')


plt.xlim(-2*limit,2*limit)
plt.ylim(-limit,limit)

plt.yticks([])
plt.xticks([])

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')


plt.show()
```

# Refraction

```python
size_of_font = 25

limit=3
points = 50
sample = 2

n_1 = 1
n_2 = 1.3

theta_var_A = np.linspace(0,90,50)
theta_var_B = np.flip(theta_var_A)
theta_var = np.concatenate([theta_var_A,theta_var_B])

theta_1 = np.deg2rad(60)
theta_2 = np.arcsin(n_1*np.sin(theta_1)/n_2)

point_coord_x = 3
point_coord_y = point_coord_x/np.tan(theta_1)
#point_A = (-2,2)
#point_A_prime = (-2,-2)
#point_B = (2,2)

y_coord_prime = 2

x_coord_prime = y_coord_prime*np.sin(theta_2)


#plt.figure(figsize=(7,7))

#plt.axes().set_aspect('equal')


fig, ax = plt.subplots(figsize=(7,4))

rect = patches.Rectangle((-2*limit,-limit), 4*limit, limit, linewidth=1, edgecolor='k', facecolor='b',alpha=0.1)
ax.add_patch(rect)


#Lines
#plt.plot((-point_coord_x,0),(point_coord_y,0),color='blue',lw=3)
line1, = ax.plot([],[],lw=3,linestyle='-',color='blue')

#plt.plot((0,x_coord_prime),(0,-y_coord_prime),color='blue',lw=3)
line2, = ax.plot([],[],lw=3,linestyle='-',color='blue')

def init():
    line1.set_data([],[])
    line2.set_data([],[])
    return (line1, line2, )

def animate(i):
    theta = theta_var[i]
    theta_rad = np.deg2rad(theta)
    point_coord_x = 3
    point_coord_y = point_coord_x/np.tan(theta_rad)
    line1.set_data((-point_coord_x,p_x),(point_coord_y,0))
    line2.set_data((p_x,point_coord_x),(0,point_coord_y))
    if i == 18:
        plt.savefig('ch14-reflection_fermat1-18.pdf',transparent=True)
#    ax.axvline(x=p_x,linewidth=1,color='k',linestyle='dotted')
    return (line1, line2, line3, line4,)


ax.set_aspect('equal')

ax.axvline(x=0,linewidth=1,color='k',linestyle='dotted')



plt.xlim(-2*limit,2*limit)
plt.ylim(-limit,limit)

plt.yticks([])
plt.xticks([])

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')

plt.tight_layout()
plt.savefig('ch14-refraction_snell1.pdf', transparent=True)
plt.show()
```

# Mirror reflections

```python

size_of_font = 25

limit=3
points = 50
sample = 2

point_of_reflection_1 = np.linspace(-0.8*limit,0.5*limit,points)
point_of_reflection_2 = np.flip(point_of_reflection_1)
point_of_reflection = np.concatenate([point_of_reflection_1,point_of_reflection_2])

P_x = point_of_reflection[sample]

point_coord_x = 3
point_coord_y = 2
point_A = (-2,2)
point_A_prime = (-2,-2)
#point_B = (2,2)


#plt.figure(figsize=(7,7))

#plt.axes().set_aspect('equal')


fig, ax = plt.subplots(figsize=(10,6))

rect = patches.Rectangle((-limit,-0.1*limit), 1.8*limit, 0.1*limit, linewidth=1, edgecolor='k', facecolor='gray',alpha=0.5)
ax.add_patch(rect)

plt.plot(-point_coord_x,point_coord_y,'o',color='blue',markersize=5)
plt.plot(-point_coord_x,-point_coord_y,'o',color='blue',markersize=5)
#plt.plot(point_coord_x,point_coord_y,'o',color='orange',markersize=5)


ax.annotate("$P$", xy=(-1.1*point_coord_x,1.1*point_coord_y),fontsize=size_of_font, color='b')
ax.annotate("$P'$", xy=(-1.2*point_coord_x,-1.2*point_coord_y),fontsize=size_of_font, color='b')
#ax.annotate("$B$", xy=(1.1*point_coord_x,1.1*point_coord_y),fontsize=size_of_font, color='orange')

#ax.annotate("$\\theta_i$", xy=(P_x-0.4,0.5),fontsize=size_of_font, color='red')
#ax.annotate("$\\theta_r$", xy=(P_x+0.1,0.5),fontsize=size_of_font, color='red')



#Lines
#plt.plot((-point_coord_x,P_x),(point_coord_y,0),color='blue',lw=3)
line1, = ax.plot([],[],lw=3,linestyle='-',color='blue')

#plt.plot((P_x,point_coord_x),(0,point_coord_y),color='blue',lw=3)
line2, = ax.plot([],[],lw=3,linestyle='-',color='blue')

#plt.plot((-point_coord_x,P_x),(-point_coord_y,0),color='blue',lw=3,linestyle='dotted')
line3, = ax.plot([],[],lw=3,linestyle='dotted',color='blue')

line4, = ax.plot([],[],lw=1,linestyle='dotted',color='k')

pointB, = ax.plot([],[],'o',color='orange',markersize=25)

#Bmark,= ax.annotate("$B$", xy=([],[]),fontsize=size_of_font, color='orange')

#pointP, = ax.annotate("$P$", xy=(P_x-0.4,0.5),fontsize=size_of_font, color='red')

#plt.plot((-point_coord_x,point_coord_x),(-point_coord_y,point_coord_y),color='orange',lw=3,linestyle='dotted')


def init():
    line1.set_data([],[])
    line2.set_data([],[])
    line3.set_data([],[])
    line4.set_data([],[])
    pointB.set_data([],[])
#    Bmark.set_data([],[])
    return (line1, line2, line3, line4, pointB, )

def animate(i):
    p_x = point_of_reflection[i]
# set_data( (x1,x2),(y1,y2))
    line1.set_data((-point_coord_x,p_x),(point_coord_y,0))
    line2.set_data((p_x,2*p_x+point_coord_x),(0,point_coord_y))
    line3.set_data((-point_coord_x,p_x),(-point_coord_y,0))
    line4.set_data((p_x,p_x),(-limit,limit))
    pointB.set_data((2*p_x+point_coord_x),(point_coord_y))
#    Bmark.set_data((2*p_x+point_coord_x),(point_coord_y))
#    if i == 18:
#        plt.savefig('ch14-reflection_fermat1-18.pdf',transparent=True)
#    ax.axvline(x=p_x,linewidth=1,color='k',linestyle='dotted')
    return (line1, line2, line3, line4, pointB, )



#ax.axvline(x=P_x,linewidth=1,color='k',linestyle='dotted')


#plt.plot((x1,x2),(y1,y2)  )

ax.set_aspect('equal')


plt.xlim(-1.6*limit,2*limit)
plt.ylim(-limit,limit)

plt.yticks([])
plt.xticks([])

ax.spines['bottom'].set_position('zero')
ax.spines['bottom'].set_color('none')


ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')


anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)
#anim

plt.tight_layout()
anim.save('ch14-reflection_amination1.gif', writer='imagemagick', fps=25)
plt.show()
anim
```
