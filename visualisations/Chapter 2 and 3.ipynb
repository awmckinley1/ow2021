```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mplines

# https://e2eml.school/matplotlib_framing.html
```

```python
# Morse Potential


r=np.linspace(-0.8,10,100)
Na = 6.022E23
R = 8.314
a=2

r0=0.74

epsilon=1.5E-21

k_e = 610 # Force constant at bottom of the well
D_e = -4.8 # Well depth
a_ww = 1 # Well width

#rep = 4*epsilon*((r0/r)**12)
#attr = -4*epsilon*((r0/r)**6)
#vlj = 4*epsilon*(((r0/r)**12) - ((r0/r)**6))

def v_morse(r):
    v = D_e*(( (1-np.exp(-a_ww*(r-r0)) )**2 )) - D_e
    return v

def v_harm(r):
    v = r**2 - 2*r*r0 + r0**2 - D_e
#    v = (r - r0)**2
    return v

def v_harmb(r):
    v_harmb = 15*a*(r-r0)**2 - D_e
    return v_harmb

plt.figure(figsize=(10,7))

ax = plt.axes()

ax.set_xlabel('Interatomic separation /nm')
ax.set_ylabel('Morse potential /J')

ax.plot(r,v_morse(r), color="blue",linewidth=1.0,linestyle="-",label="Morse potential")

ax.plot(r,v_harm(r), color="black",linewidth=1.0,linestyle="-",label="Harmonic oscillator")
ax.plot(r,v_harmb(r), color="blue",linewidth=1.0,linestyle="-",label="Harmonic oscillator")




#plot(r,attr, color="red",linewidth=1.0,linestyle="-",label="Attraction $ \propto -{1}/{r^6}$")
#plot(r,vlj, color="green",linewidth=1.0,linestyle="-",label="Lennard Jones potential")
#plot(s,fs3, color="red",linewidth=1.0,linestyle="-",label="M = 71 g mol$^{-1}$")

ax.legend(loc='upper right')

#get the axis object
#ax = gca()
#make the right spine invisible
ax.spines['right'].set_color('none')
#make the top spine invisible
ax.spines['top'].set_color('none')
#move the bottom spine to the zero of y axis
ax.spines['bottom'].set_position(['data',-1.2*D_e])
#move the left spine to the zero of x axis
#ax.spines['left'].set_position(['data',0])
#keep ticks only on spines that remain visible
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.axhline(y=0,linewidth=0.5,color='k',linestyle="--")
ax.set_ylim(-1.2*D_e, 1.2*D_e)




#ylim(-2E-21,2E-21)

#savefig('morse_pot_1.pdf')

plt.show()
```
