```python
%config InlineBackend.figure_format = 'retina'
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mplines
import matplotlib
import matplotlib.patches as patches


from matplotlib import animation, rc
from IPython.display import HTML, Image

rc('animation', html='html5') # Needed for in-line movie
```

# Animating reflection refraction

```python

size_of_font = 15

limit=3
points = 10
sample = 2

point_of_reflection_1 = np.linspace(-limit,limit,points)
point_of_reflection_2 = np.flip(point_of_reflection_1)
point_of_reflection = np.concatenate([point_of_reflection_1,point_of_reflection_2])

P_x = point_of_reflection[sample]

point_coord_x = 3
point_coord_y = 2
point_A = (-2,2)
point_A_prime = (-2,-2)
point_B = (2,2)


plt.figure(figsize=(10,10))

#plt.axes().set_aspect('equal')


ax = plt.axes()

rect = patches.Rectangle((-2*limit,-limit), 4*limit, limit, linewidth=1, edgecolor='k', facecolor='b',alpha=0.1)
ax.add_patch(rect)

plt.plot(-point_coord_x,point_coord_y,'o',color='blue',markersize=5)
plt.plot(-point_coord_x,-point_coord_y,'o',color='blue',markersize=5)
plt.plot(point_coord_x,point_coord_y,'o',color='orange',markersize=5)


ax.annotate("$A$", xy=(-1.1*point_coord_x,1.1*point_coord_y),fontsize=size_of_font, color='b')
ax.annotate("$A'$", xy=(-1.2*point_coord_x,-1.2*point_coord_y),fontsize=size_of_font, color='b')
ax.annotate("$B$", xy=(1.1*point_coord_x,1.1*point_coord_y),fontsize=size_of_font, color='orange')

ax.annotate("$\\theta_i$", xy=(P_x-0.4,0.5),fontsize=size_of_font, color='red')
ax.annotate("$\\theta_r$", xy=(P_x+0.1,0.5),fontsize=size_of_font, color='red')




plt.plot((-point_coord_x,P_x),(point_coord_y,0),color='blue',lw=3)
plt.plot((P_x,point_coord_x),(0,point_coord_y),color='blue',lw=3)
plt.plot((-point_coord_x,P_x),(-point_coord_y,0),color='blue',lw=3,linestyle='dotted')
plt.plot((-point_coord_x,point_coord_x),(-point_coord_y,point_coord_y),color='orange',lw=3,linestyle='dotted')



ax.axvline(x=P_x,linewidth=1,color='k',linestyle='dotted')


#plt.plot((x1,x2),(y1,y2)  )

ax.set_aspect('equal')


plt.xlim(-2*limit,2*limit)
plt.ylim(-limit,limit)

plt.yticks([])
plt.xticks([])

ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')


plt.show()
```
